# DATA PREPROCESSING



*   Remove unnecessary columns: Drop "Restaurant ID", "Address", "Locality", "Locality Verbose", "Rating color", and "Rating text".
*   Handle missing values: Fill missing values in "Cuisines" with "Unknown".
*   Encode categorical variables: Convert "City", "Cuisines", "Currency", and binary columns to numerical form.
*   Split into training & testing sets.




```python
#remove unnecessary columns
data=data.drop(columns=["Restaurant ID","Address", "Locality","Locality Verbose","Rating color","Rating text"],axis=1,errors='ignore')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(y=data['Cuisines'], order=data['Cuisines'].value_counts().index[:10], palette="coolwarm")
plt.xlabel("Count")
plt.ylabel("Cuisines")
plt.title("Top 10 Most Frequent Cuisines")
plt.show()
```

```python
#now we can fill this cuisines missing vaalues with the use of most frequant cuisines so it is more relevent
data["Cuisines"].fillna(data["Cuisines"].mode()[0],inplace=True)
```

label encoding

```python
data['Has Online delivery'] = data['Has Online delivery'].map({'Yes': 1,'No':0})
data['Has Table booking'] = data['Has Table booking'].map({'Yes': 1,'No': 0})
data['Is delivering now'] = data['Is delivering now'].map({'Yes': 1,'No': 0})
```

```python
data = pd.get_dummies(data, columns=['City','Cuisines','Currency'],drop_first=True , dtype=int)
```

```python
data['Switch to order menu'] = data['Switch to order menu'].map({'Yes': 1,'No':0})
```

data split into

```python
#remove restaurat name
data=data.drop(columns="Restaurant Name",axis=1)
```

**now we drop rating colum for feature prediction**

```python
x = data.drop(columns='Aggregate rating' , axis=1)
y = data['Aggregate rating']
```

# train and test

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

# import model

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score
model = DecisionTreeRegressor(random_state=42)

model.fit(X_train, y_train)
y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R² Score: {r2}")
```
